```python
import os
from mario_gpt.utils import convert_level_to_png
```

```python
folder_path = 'experiment_levels/experiment_levels_5/generated_files_(Y-Y-N)'

files = os.listdir(folder_path)
for file_name in files:
    if file_name.endswith('.txt'):
        file_path = os.path.join(folder_path, file_name)

        with open(file_path, 'r') as file:
            file_content = file.read()
            image_data = convert_level_to_png(file_content.split('\n'))[0] # file_content.split('\n')[:-1] needed if inspecting levels from VGLC
            display(image_data)
```
